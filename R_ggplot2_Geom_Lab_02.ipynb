# ggplot2
## Geom_Bar()
geom_bar() geom_col() stat_count()
Bar charts

There are two types of bar charts: geom_bar() and geom_col(). geom_bar() makes the height of the bar proportional to the number of cases in each group (or if the weight aesthetic is supplied, the sum of the weights). If you want the heights of the bars to represent values in the data, use geom_col() instead. geom_bar() uses stat_count() by default: it counts the number of cases at each x position. geom_col() uses stat_identity(): it leaves the data as is.



```
geom_bar(
  mapping = NULL,
  data = NULL,
  stat = "count",
  position = "stack",
  ...,
  just = 0.5,
  width = NULL,
  na.rm = FALSE,
  orientation = NA,
  show.legend = NA,
  inherit.aes = TRUE
)

geom_col(
  mapping = NULL,
  data = NULL,
  position = "stack",
  ...,
  just = 0.5,
  width = NULL,
  na.rm = FALSE,
  show.legend = NA,
  inherit.aes = TRUE
)

stat_count(
  mapping = NULL,
  data = NULL,
  geom = "bar",
  position = "stack",
  ...,
  width = NULL,
  na.rm = FALSE,
  orientation = NA,
  show.legend = NA,
  inherit.aes = TRUE
)
```

##### Geom_bar is designed to make it easy to create bar charts that show

````r
 ggplot(mpg,aes(class))+geom_bar()
````

````r
df =ggplot(mpg,aes(class))
df+geom_bar()
````

###### Total engine displacement of each class

````r
df + geom_bar(aes(weight = displ))
````

#####  Class to y instead to flip the orientation

````r
ggplot(mpg) + geom_bar(aes(y = class))
````

#####  Bar charts are automatically stacked when multiple bars are placed at the same location. The order of the fill is designed to match the legend

````r
df + geom_bar(aes(fill = drv))
````

##### If you need to flip the order (because you've flipped the orientation)call position_stack() explicitly:

````r
ggplot(mpg, aes(y = class)) +
 geom_bar(aes(fill = drv), position = position_stack(reverse = TRUE)) +
 theme(legend.position = "top")
````
