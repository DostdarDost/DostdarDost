## (Geom_plots)
### 1.  Geom_lines: Horizontal, Vertical, and Diagonal

These geoms add reference lines (sometimes called rules) to a plot, either horizontal, vertical, or diagonal (specified by slope and intercept). These are useful for annotating plots.

```r
library(ggplot2)

# Create a data frame
df <- data.frame(x = c(-4, -3, -2, -1, 0, 1, 2, 3, 4),
                 y = c(16, 9, 4, 1, 0, 1, 4, 9, 16 ))
# Create a scatter plot with multiple horizontal lines
ggplot(df, aes(x = x, y = y)) +
  geom_point(color="blue") +
  geom_vline(xintercept = c(-3, 2, 0), linetype = c("solid", "dashed", "dotted"), color = c("black", "red", "green"), size = c(1, 2, 1.5)) + # Add the + operator here to include the title and labels as layers
  ggtitle("Scatter Plot with Horizontal Line") +
  xlab("X-axis") +
  ylab("Y-axis")
```
